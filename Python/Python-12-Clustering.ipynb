```python
#Clustering (잘 사용하지 않는 것이 좋다.)

import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt

import warnings #간단한 오류는 무시하게 하는 명령어
warnings.filterwarnings('ignore')

import os
from os.path import join
```

```python
from sklearn.datasets import load_wine
wine = load_wine()

print(wine.DESCR)
```

```python
data = wine.data
label = wine.target
columns = wine.feature_names

data = pd.DataFrame(data, columns = columns )
data.head()
```

```python
#데이터 전처리 1.스케일링, 2.차원축소(값들이 차이가 나면 차원이 생김)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
data[:5]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) #따로 언급 없으면 2차원으로 축소하는것이 관례
data = pca.fit_transform(data)
data[:5] 
```

```python
data.shape #13차원에서 2차원으로 축소된 것 확인
```

```python
# K-Means 모델의 활용
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)

kmeans.fit(data)
```

```python
cluster = kmeans.predict(data)
cluster
```

```python
plt.scatter(data[:,0], data[:,1], c=cluster, linewidth=1, edgecolor='black')
```

```python
from scipy.cluster.hierarchy import single
# Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering
single_clustering = AgglomerativeClustering(n_clusters=3, linkage='single') #3개로 나눠라.

single_clustering.fit(data)
single_cluster = single_clustering.labels_
single_cluster
```

```python
# 산점도
plt.scatter(data[:,0],data[:,1], c=single_cluster)
```

```python
complete_clustering = AgglomerativeClustering(n_clusters=3, linkage='complete')
complete_clustering.fit(data)
complete_cluster = complete_clustering.labels_
complete_cluster
```

```python
plt.scatter(data[:,0],data[:,1], c=complete_cluster)
```

```python
# 덴드로그램(dendrogram)
from scipy.cluster.hierarchy import dendrogram

#Hierarchical Cluster의 자식 노드
children = complete_clustering.children_

distance = np.arange(children.shape[0])

# 각 클러스터 단계를 포함한 노드의 수 계산
no_of_observations = np.arange(2, children.shape[0]+2) # 2추가하는 이유? 인덱스가 부족/여기서는 최상위 노드까지 필요해서 1+1 2개 필요

linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

# 덴드로그램을 그린다.
dendrogram(linkage_matrix, p = len(data), labels = complete_cluster, show_contracted=True, no_labels=True)
```

```python
# 가장 좋은 클러스터를 형성하는 클러스터의 수를 찾아보자
from sklearn.metrics import silhouette_score

best_n = -1   #-1로 초기화하는게 관행
best_score = -1

for n_cluster in range(2, 11):
  kmeans = KMeans(n_clusters=n_cluster)
  kmeans.fit(data)
  cluster = kmeans.predict(data)

  #실루엣 점수를 본다. 
  score = silhouette_score(data, cluster)

  print('Cluster count: {} Silhouette Score: {:.2f}'.format(n_cluster, score))

  if score > best_score:
    best_n = n_cluster
    best_score = score


print('\n--------------------------------------------------------\n')
print('Best n_cluster: {} Silhoutette Score: {:.2f}'.format(best_n, best_score))

  #결과값 - 실루엣 점수가 높을 수록 좋다.
```
