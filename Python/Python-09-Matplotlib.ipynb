```python
# 한번에 두 개의 그래프를 출력하는 방법
plt.figure()   # :그림판 정의
```

```python
plt.subplot(2, 1, 1)     #그림판의 위치표시: rows, columns, panel number
plt.plot(x,np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x,np.cos(x))
```

```python
plt.plot(x, np.sin(x), color = 'blue')
plt.plot(x, np.sin(x-1),color = 'g')
plt.plot(x, np.sin(x-2), color = '0.75') #0은 흰색, 1은 검은색 
plt.plot(x, np.sin(x-3), color = '#FFDD44') #16진수로 표현
plt.plot(x, np.sin(x-4), color = (1.0, 0.2, 0.3)) #RGB로 표현
plt.plot(x, np.sin(x-5), color = 'chartreuse') #색상의 이름 거의 사용x
```

```python
plt.plot(x,x+0, linestyle='solid')
plt.plot(x,x+1, linestyle='dashed')
plt.plot(x,x+2 ,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted')

plt.plot(x,x+0, linestyle='-')
plt.plot(x,x+4, linestyle='--')
plt.plot(x,x+5 ,linestyle='-.')
plt.plot(x,x+6,linestyle=':')
```

```python
plt.plot(x, x+0, '-g')    #solid green
plt.plot(x, x+1 , '--c')  #dashed green
plt.plot(x, x+2, '-.k')   #dashdot black
plt.plot(x, x+3, ':r')    #dotted red
```

```python
# 세부 조정하는 법

plt.plot(x, np.sin(x)) #아무런 옵션을 주지 않으면 화면 꽉채움
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10,0)
plt.ylim(-1.2, -1.2)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')   #타이트하게 출력, 기본디폴트값
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal') #x,y비율을 동일하게 준다.
```

```python
#label 붙이기
plt.plot(x,np.sin(x))
plt.title('A Sine Curve') #타이틀 입력
plt.xlabel('x')  #X축 제목 입력
plt.ylabel('sin(x)') #Y축 제목 입력
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x),':b', label = 'cos(x)')
plt.axis('equal')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin and Cos')
```

```python
#산점도의 표현 (Scatter Plot)
x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x, y,'o', color='black' )  #marker 'o'
```

```python
#랜덤으로 점 찍기
np.random.RandomState(0)
```

```python
rng = np.random.RandomState(0) #seed값 0
for marker in ['o',',',',','x','+','v','^','<','>','s','d']:
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
  plt.legend(numpoints=1) #기본값이 2개라 1로 1개 표현
  plt.xlim(0,1.8)
```

```python
plt.plot(x, y,'-p', color='gray',
         markersize=15,
         linewidth=4,
         markerfacecolor='white',
         markeredgecolor='red',
         markeredgewidth=2
)
```

```python
# plt.scatter 산점도를 표현

rng = np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)

plt.scatter(x, y)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0],features[1], c=iris.target,s=features[3]*100,alpha=0.2)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
iris.feature_names[3]
```
